# Problem 1

#### a.) Create a code that asks a user to enter 10 names that will be added to a list

```python
list_of_names = [ ] # Initialize an empty array 
n = 10 #Let n be the number of iterations 
  
for i in range(0, n): #Create a for loop 
    names = input("Enter a name: ") #Take input
    list_of_names.append(names) #Append input to the variable list of names
      
print(list_of_names) #Print list of names
```

#### b.) Split the list of 10 names into three other lists, a list of short names containing less than four characters, a list of four character names and a list of long names with more than four characters

```python
#Initialize empty arrays
less_than_four = []
equal_four = []
more_than_four = []

for i in list_of_names: #Create for loop
    #conditional statements
    if len(i) < 4:  
        less_than_four.append(i) #If true, then append the element to the corresponding variable name
    elif len(i) == 4:
        equal_four.append(i) #If true, then append the element to the corresponding variable name
    else: #Otherwise, if both the prior conditions are false
        more_than_four.append(i) #then append the element to the corresponding variable name
        
print('Short Names: %s' %less_than_four,
      'Four Characters Names: %s' %equal_four, 'Long Names: %s' %more_than_four) #Print the result
```

#### c.) Create a new list that starts with the first name and contains every second name, using the ":"-notation for indexation. 

```python
#Create new list that contains the first element and the every other element of list of names
new_list = list_of_names[:len(list_of_names):2] #Start 0, Stop until the lenght of elements in the list, Step 2
print(new_list) #Print new list
```

#### d.) Create a new list that is equal to the list from a., except that it starts with the last name and goes backwards, using the ":"-notation. 

```python
reverse_list_of_names = list_of_names[::-1] #Start at 0, Stop at 0, Step in reverse
print(reverse_list_of_names) # Print Reverse list of names
```

#### e.) Create two lists of integers, numberlist1 and numberlist2, each with 8 elements. Then, create a third list of numbers, that will take the value equal to the sum of the (corresponding) elements from numberlist1 and numberlist2. If they are equal, the new element should have a value equal to their product. If the element from numberlist2 is bigger, then the new element should be the difference between the element from numberlist2 and numberlist1. 

```python
#Create list of integers
numlist1 = [0,5,2,5,21,34,4,7]
numlist2 = [1,5,3,8,13,55,9,7]
numlist3 = [0,0,0,0,0,0,0,0] #Placeholder #Symmetrical 

for i in range(0,8): #For loop
    
    #Conditions
    if numlist1[i] == numlist2[i]:
        numlist3[i] = numlist1[i] * numlist2[i]
        
    elif numlist2[i] > numlist1[i]:
        numlist3[i] = numlist2[i] - numlist1[i]
        
    else:
        numlist3[i] = numlist1[i] + numlist2[i]
        
print(numlist3) #Print
```

#### a.) Create an nd-array taking on lots of values from -2 to 2. Create a new variable that is a third order polynomial in this original expression. 

```python
x_val = np.linspace(-2, 2,1000)
y_val = (x_val ** 3) + (x_val ** 2) - (x_val) + 1
```

#### b.) Plot the function and its numerical derivative (as from Problem 3e.) Show that any local maxima or minima corresponds to the derivative being equal to zero. (And adjust your coefficients in 4a so that you will have a local maximum and minimum if necessary!)

```python
#Numerical Derivative 
change_in_y_val = np.diff(y_val)
change_in_x_val = np.diff(x_val)

num_dev = change_in_y_val / change_in_x_val

plt.plot(y_val, '-r', label = "Third Order Polynomials")
plt.plot(num_dev, '-b', label = "Numerical Derivative of the Polynomial")


plt.axhline(1.75, color = 'g', linestyle = '--', label = 'roots')
plt.axvline(625, color = 'g', linestyle = '--')

plt.legend(loc = "upper right")
```

#### a.) Make a nice-looking figure of a partial equilibrium model in microeconomics, that is a market cross. Make sure that you label the axes and the demand and supply curves and create dotted lines that show the equilibrium price and quantity on the axes. 

```python
#Suppose 
a = 200
b = 50
c = 80
d = 30

Q = np.linspace(50,250,1000)

#Inverse Demand Function 
price_d = (a - Q)/b

#Supply Function 
price_s = (Q - c)/d

#Price Equilibrium 
pe = (a-c)/(d+b)

print('Equilibrium Price:', pe)

plt.plot(price_d, 'b', label = "Inverse Demand Function")
plt.plot(price_s, 'r', label = "Inverse Supply function")
plt.axhline(pe, color = 'g', linestyle = '--', label = 'Equilibrium Price and Quantity')
plt.axvline(375, color = 'g', linestyle = '--')
plt.plot(375,1.5,'o', markersize = 10)
plt.title("Microeconomic Equilibrium")
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.legend(loc = "upper right")
```
